## Problem statement
Woodler is a locall company that prepares and distributes wood to be used in building and carpentry.
They have one large distribution center (DC) and many stores. Stores are located in the city and don't have much storage capacity. So, Woodler also own many trucks that they use to fufill the stores in a daily basis. Because demand is vary volatile (it varies with the weather, school time off, etc.), they often have to rent extra trucks to deliver daily loads. 

Renting extra trucks is expensive though. But to avoid renting trucks, Woodler must plan their distribution very well. Which is not an easy task because trucks are small, so they can cirlulate in the city. In addition, timber is a heavy load and takes up a lot of space.

Because traffic is heavy, each truck leaves the DC in the morning and only returns in the afternoon, not having time for a second trip.

Placement of orders from the stores are closed by 5PM of previous day. This means that Woodler knows in advance exacly how much to deliver to each store. So the goal is to deliver the demand of every store with the minimum number of trucks.

## Formulation


### Input Data Model

#### Set of indices
- $I$ set.

#### Parameters
* $d_i$ demand of site $i$.

### Decision Variables


### Constraints

* Must visit every store once:
$$\sum_{ik} x_{ijk} = 1, \quad \forall j.$$

* Every truck departs from the DC:
$$\sum_j x_{0jk} \leq nu_j, \quad \forall j.$$

### Objective
The objective is to minimize the total cost of the diet.
$$\min{\sum_i c_ix_i}.$$

### Final formulation
$$
\begin{eqnarray}
\begin{array}{rcl}
& \min & \sum_i c_ix_i\\
& \text{s.t.}& \sum_i nq_{ij}x_i \geq nl_j, \quad \forall j,\\
&& \sum_i nq_{ij}x_i \leq nu_j, \quad \forall j,\\
&& x_i \geq 0, \quad \forall i.
\end{array}
\end{eqnarray}
$$

### Delivery Cost
In the base model, each food has a per unit cost. Now suppose that, in addition, there is a delivery cost (fixed cost) associated with each food.

#### Parameters
* $cd_i$ delivery cost of food $i$.
* $fu_i$ upper bound on the quantity of food $i$ that can be used in the diet.

#### Decision variables:
* $z_i$	equals $1$ if food $i$ is part of the diet, $0$ otherwise.

#### Constraints 
* If food $i$ is part of the diet, then $z_i$ must be $1$:
$$x_i \leq fu_i z_i, \quad \forall i.$$
Observe that this is a classic *if-then* constraint and it uses a *big-M*. If you are not familiar with these two concepts, make sure to have a look at the [Ukulele-la-la](https://github.com/mip-master/learning_mip/tree/master/ukulelelala) use case.

#### Objective:
The following term must be added to the objective function to ensure that the delivery cost gets charged for every selected food.
$$\sum_i cd_iz_i$$

Observe that $fu_i$ can be derived from the max-nutrients constraint. In fact, for every food $i$ we have the following:
$$
nq_{ij} x_i \leq nu_j \forall j 
\Rightarrow x_i \leq \frac{nu_j}{nq_{ij}} \forall j 
\Rightarrow x_i \leq \max_j{\frac{nu_j}{nq_{ij}}}.
$$
Therefore, we can define:
$$fu_i := \max_j{\frac{nu_j}{nq_{ij}}}.$$
