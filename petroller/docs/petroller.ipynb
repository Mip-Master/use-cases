# Petroller



## Formulation


### Input Data Model

#### Set of indices
- $I$ set of crude oil sites
- $J$ set of derived fuels

#### Parameters
- $fp_{ij}$ yield (fraction) of fuel $j$ from crude oil coming from site $i$.
- $a_{i}$ maximum availability of crude oil from site $i$.
- $d_{j}$ demand for fuel $j$.
- $r_{j}$ revenue from fuel $j$.
- $cp_{i}$ procurement cost of crude oil coming from site $i$.
- $cs_{j}$ storage cost of fuel $j$.
- $dp_{j}$ penalty (per barrel) for not meeting demand of fuel $j$.

### Decision variables
- $x_{ij}$ amount (barrels) of crude oil from site $i$ used to produce fuel $j$.
- $y_{j}$ amount (barrels) of fuel $j$ produced from crude oil.
- $s_{j}$ amount (barrels) of fuel $j$ stored.
- $z_{j}$ amount (barrels) of fuel $j$ outsourced.

### Constraints

- *Oil availability:*
$$
\sum_{j} x_{ij} \leq a_i, \quad \forall i.
$$

- *Yield of fuel:*
$$
y_j = \sum_{i} fp_{ij} \cdot x_{ij}, \quad \forall j.
$$
- *Flow balance:*
$$
y_j + z_j = s_j + d_j , \quad \forall j.
$$ 

### Objective function

The objective is to maximize the total revenue and minimize the total procurement cost, storage cost, and unmet demand penalty.

#### Revenue
$$
revenue = \sum_{j}  r_j \cdot y_j.
$$

#### Procurement cost
$$
procurement = \sum_{i}  cp_i \cdot x_{ij}.
$$

#### Storage cost
$$
storage = \sum_{j}  cs_j \cdot s_j.
$$

#### Unmet demand penalty
$$
penalty = \sum_{j}  dp_j \cdot z_j.
$$


#### Final objective
Finally, the objective function is the following:
$$\max revenue - procurement - storage - penalty,$$
where $revenue, procurement, storage$ and $penalty$ are defined above.
