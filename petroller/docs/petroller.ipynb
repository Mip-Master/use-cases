# Petroller

Petroller owns an petroleum refinery plant that produce gasoline, diesel, and jet fuel from crude oil. They can procure crude oil from multiple oil production plants (which is not a refinery), but it can differ significantly in term of quality and price. For example, a particular oil production plant might offer crude oil at a low price but the yield of fuel that can be derived from it might be low as well. In addition, the quantity of crude oil from each source is limited and must be taken into account in the operations.

Petroller is expected to meet a minimum demand for each type of fuel, and a penalty inccurs for each unmet unit. Supplying more than the the minimum demand is acceptable.


## Formulation
This problem can be modeled as linear program.

### Input Data Model

#### Set of indices
- $I$ set of crude oil production plants
- $J$ set of derived fuels

#### Parameters
- $r_{ij}$ yield (ratio) of fuel $j$ from crude oil coming from site $i$.
- $a_{i}$ maximum availability of crude oil from site $i$.
- $d_{j}$ demand for fuel $j$.
- $rv_{j}$ revenue from fuel $j$.
- $cp_{i}$ procurement cost of crude oil coming from site $i$.
- $dp_{j}$ penalty (per barrel) for not meeting demand of fuel $j$.

### Decision variables
- $x_{ij}$ amount (barrels) of crude oil from site $i$ used to produce fuel $j$.
- $y_{j}$ amount (barrels) of fuel $j$ produced from crude oil.
- $z_{j}$ amount (barrels) of fuel $j$ outsourced.

### Constraints

- *Oil availability:*
$$
\sum_{j} x_{ij} \leq a_i, \quad \forall i.
$$

- *Yield of fuel:*
$$
y_j = \sum_{i} r_{ij} \cdot x_{ij}, \quad \forall j.
$$
- *Flow balance:*
$$
y_j + z_j \geq d_j , \quad \forall j.
$$ 

### Objective function

The objective is to maximize the total revenue and minimize the total procurement cost and unmet demand penalty.

#### Revenue
$$
revenue = \sum_{j}  rv_j \cdot y_j.
$$

#### Procurement cost
$$
procurement = \sum_{i}  cp_i \cdot x_{ij}.
$$

#### Unmet demand penalty
$$
penalty = \sum_{j}  dp_j \cdot z_j.
$$


#### Final objective
Finally, the objective function is the following:
$$\max revenue - procurement - penalty,$$
where $revenue, procurement$ and $penalty$ are defined above.
